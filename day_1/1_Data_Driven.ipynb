# 1. Data Driven

Data Driven (o basado en datos) es una metodología para tomar decisiones y definir estrategias, adaptar tácticas y controlar operaciones. En este enfoque, la recolección y el análisis de datos son fundamentales. Por lo general se caracteriza por grandes volúmenes de datos.  

 ##   1.1. ¿Qué es Data Driven?


- _Sensus lato_ Data Driven Decision-Making (DDDM) se refiere a tomar decisiones basadas en información sólida extraída de datos.

- Para aplicar DDDM las organizaciones van más allá de las intuiciones o conjeturas, utilizan pruebas y análisis de datos para guiar sus acciones.

- La cultura Data Driven esta mayormente asociada a entornos de empresas de tecnología que por sus características manejan grandes cantidades de datos. Sin embargo, la cultura de DDDM impulsado por el acceso a datos se encuentra en expansión hacia todo tipo de organizaciones y emprendimientos.






 ##   1.2. Ventajas del Toma de Decisiones Basada en Datos (DDDM)



- Precisión: Las decisiones basadas en datos son más precisas y fundamentadas.

- Identificación de tendencias: Permite detectar rápidamente tendencias y ajustar estrategias.

- Seguimiento de métricas: Facilita el seguimiento de indicadores de rendimiento

**El descubrimiento basado en datos en la ciencia**

Data-driven discovery in science se refiere al proceso de descubrir nuevas ideas, patrones y relaciones a partir de grandes conjuntos de datos utilizando técnicas computacionales.

En la última década, el big data y la inteligencia artificial (IA) han alterado significativamente los estilos de vida y la prosperidad general de la sociedad, al tiempo que han influido en un cuarto paradigma científico:

> la ciencia intensiva en datos o basada en datos.

Este cuarto paradigma sigue los paradigmas tradicionales de las ciencias

- experimentales,

- teóricas y

- computacionales (Bell et al. 2009; Hey et al. 2009).

Impulsados por conjuntos de datos en constante expansión y armados con tecnologías digitales (por ejemplo, IA, análisis de big data y supercomputación), los geocientíficos están avanzando en los enfoques tradicionales de pensamiento tanto en la industria como en la academia (Bergen et al. 2019; Sun et al. 2022).


Ver: [The Fourth Paradigm – Data-Intensive Scientific Discovery](https://link.springer.com/chapter/10.1007/978-3-642-33299-9_1)





_Data-driven discovery in science_ en algunos dominios cientificos

**Geociencia:**

- Los geocientíficos estudian la Tierra sólida, los océanos y la atmósfera como un sistema complejo e interactivo.

- Los procesos abarcan desde escalas atómicas hasta planetarias, y comprenderlos requiere analizar grandes conjuntos de datos geológicos y simulaciones computacionalmente intensivas.

- El aprendizaje automático (machine learning, ML) desempeña un papel crucial en la geociencia al automatizar tareas complejas, modelar relaciones y revelar nuevos patrones.

- Por ejemplo, ML se puede utilizar para la cartografía geológica, la modelización de sistemas de fracturas y la clasificación de partículas de ceniza volcánica.

**Algoritmos para el modelado basado en datos, el descubrimiento de leyes y el aprendizaje de restricciones físicas.**

- Una limitación clave de las e deep neural networks y otros métodos similares basados en datos es la falta de interpretabilidad del modelo resultante: se centran en la predicción y no proporcionan ecuaciones de gobierno ni modelos claramente interpretables en términos del conjunto de variables original.

- Un enfoque alternativo basado en datos utiliza la regresión simbólica para identificar directamente la estructura de un sistema dinámico no lineal a partir de datos

- Sin ningún conocimiento previo sobre física, cinemática o geometría, el algoritmo propuesto por MICHAEL SCHMIDT Y HOD LIPSON descubrió hamiltonianos, lagrangianos y otras leyes de conservación del momento y geométrica.

- La tasa de descubrimiento se aceleró a medida que las leyes encontradas para sistemas más simples se utilizaron para iniciar explicaciones para sistemas más complejos, descubriendo gradualmente el "alfabeto" utilizado para describir esos sistemas.
